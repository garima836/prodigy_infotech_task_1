```python
#Import Required Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the Dataset
df = pd.read_csv('API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv', skiprows=4)
```

```python
# Inspect the Dataset
print(df.head())
```

```python
#Create a Bar Chart (Categorical Variable Example)
# Filter top 10 countries by 2023 population
top_countries = df[['Country Name', '2023']].sort_values(by='2023', ascending=False).head(10)

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x='2023', y='Country Name', data=top_countries, palette='viridis')
plt.title('Top 10 Countries by Population (2023)')
plt.xlabel('Population')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

```python
#Create a Histogram (Continuous Variable Example)
# Drop NaN values and plot histogram
plt.figure(figsize=(8, 5))
sns.histplot(df['2023'].dropna(), bins=30, kde=True)
plt.title('Distribution of Population Across Countries (2023)')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.tight_layout()
plt.show()
```
